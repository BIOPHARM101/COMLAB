## Q2 (5 points)
จงเขียนโปรแกรมที่รับรหัสนิสิต และ ภาคการศึกษา (1/2/3) แล้วแสดงค่าธรรมเนียมการศึกษาของนิสิตในภาค <br>
นั้นตามตารางข้างล่างนี้ซึ่ง เก็บเป็น **dictionary** <br>



เลขหลักที่ 3 (จากหน้า) ของรหัสนิสิตระดับปริญญาบัณฑิต คือ 3, 4 และรหัสนิสิตระดับบัณฑิตศึกษา เป็น 7 <br>
เลข 2 หลักสุดท้ายของรหัสนิสิตเป็นรหัสคณะ <br>
รหัสคณะในกลุ่ม 1 คือ 21 23 25 28 30 31 32 33 35 36 37 38 39 53  <br>
รหัสคณะในกลุ่ม 2 คือ 22 24 26 27 29 34 40 51  <br>
และแสดงผลลัพธ์ดังตัวอย่างข้างล่างนี้ <br>

จงเขียนฟังก์ชัน `get_dict()` ที่สร้าง dictionary จากตารางค่าเทอม ซึ่งมี <br>

- key เป็นสตริงของรหัสคณะในกลุ่ม 1 และ 2 มีค่าเป็น `group1` หรือ `group2`
- value เป็น dictionary ที่มี 
    - key เป็นสตริงของระดับการศึกษา มีค่าเป็น `Bachelor` หรือ `Graduate`
    - value เป็น dictionary ที่มี
        - key เป็นสตริงของภาคการศึกษา ภาค1/2 มีค่าเป็น `S1_2` หรือ ภาค 3 มีค่าเป็น `S3`
        - value เป็น dictionary ที่มี
            - key เป็นสตริงของรหัสนิสิต มีค่าเป็น `ID_1` , `ID_2` หรือ `ID_3` สำหรับรหัสนิสิตที่ขึ้นต้นด้วยปี 48-49 , 50-55 และ 56 เป็นต้นไป ตามลำดับ
            - value เป็นค่าเทอม

และแสดงผลลัพธ์เป็นข้อมูลใน dictionary ตามตัวอย่างการทำงาน


### ตัวอย่างการทำงาน: 

Enter student ID : **6033235423**<br>
Enter semester : **1**<br>
Registration fee : 21000

### ตัวอย่างการทำงาน:

Enter semester : **2**<br>
Registration fee : 31000

### ตัวอย่างการทำงาน:

Enter semester : **3**<br>
Registration fee : 7000

### ตัวอย่างการทำงาน:

Enter semester : **0**<br>
Invalid semester

### ตัวอย่างการทำงาน:

Invalid ID

### ตัวอย่างการทำงาน:

Invalid ID

```python
def get_dict():
    data = {
        'group1':{
            'Bachelor':{
                'S1_2':{'ID_1': 16000, 'ID_2': 18000, 'ID_3': 21000},
                'S3': {'ID_1': 4000, 'ID_2': 4500, 'ID_3': 5250}},
            'Graduate':{
                'S1_2':{'ID_1': 22500, 'ID_2': 26000, 'ID_3': 31000}, 
                'S3': {'ID_1': 6000, 'ID_2': 7000, 'ID_3': 7750}}},
        'group2':{
            'Bachelor':{
                'S1_2':{'ID_1': 12000, 'ID_2': 14500, 'ID_3': 17000}},
                'S3': {'ID_1': 4000, 'ID_2': 4500, 'ID_3': 5250},
            'Graduate': {
                'S1_2': {'ID_1': 16500, 'ID_2': 19000, 'ID_3': 23000}, 
                'S3': {'ID_1': 6000, 'ID_2': 7000, 'ID_3': 7750}}} }
    return data
def run():
    '''รับเลขประจำตัวนิสิต ตรวจสอบ และแสดงค่าเทอม'''
    data= get_dict()
    in_st  = input()
    if len(in_st) == 10 and int(in_st[0:2]) >= 48 and int(in_st[0:2])<= 65:
        in_num = input()
        if int(in_st[8:10]) in [21,23,25,28,30,31,32,33,35,36,37,38,39,53]: #group1
            if int(in_st[0:2]) in [48,49]: #-------- row1 group1
                if int(in_st[2]) in [3,4]:
                    if int(in_num) in [1,2]:
                        print('Registration fee :',data['group1']['Bachelor']['S1_2']['ID_1'])
                    elif int(in_num) == 3:
                        print('Registration fee :',data['group1']['Bachelor']['S3']['ID_1'])
                    else:print('Invalid semester')
                elif int(in_st[2]) == 7 : 
                    if int(in_num) in [1,2]:
                        print('Registration fee :',data['group1']['Graduate']['S1_2']['ID_1'])
                    elif int(in_num) == 3:
                        print('Registration fee :',data['group1']['Graduate']['S3']['ID_1'])
                    else:print('Invalid semester')
                    

            elif int(in_st[0:2]) in [50,51,52,53,54,55]: # row2 group1
                if int(in_st[2]) in [3,4]:
                    if int(in_num) in [1,2]:
                        print('Registration fee :',data['group1']['Bachelor']['S1_2']['ID_2'])
                    elif int(in_num) == 3:
                        print('Registration fee :',data['group1']['Bachelor']['S3']['ID_2'])
                    else:print('Invalid semester')
                elif int(in_st[2]) == 7 : 
                    if int(in_num) in [1,2]:
                        print('Registration fee :',data['group1']['Graduate']['S1_2']['ID_2'])
                    elif int(in_num) == 3:
                        print('Registration fee :',data['group1']['Graduate']['S3']['ID_2'])
                    else:print('Invalid semester')

            elif int(in_st[0:2]) in [56,57,58,59,60,61,62,63,64,65]: # row3 group1
                if int(in_st[2]) in [3,4]:
                    if int(in_num) in [1,2]:
                        print('Registration fee :',data['group1']['Bachelor']['S1_2']['ID_3'])
                    elif int(in_num) == 3:
                        print('Registration fee :',data['group1']['Bachelor']['S3']['ID_3'])
                    else:print('Invalid semester')
                elif int(in_st[2]) == 7 : 
                    if int(in_num) in [1,2]:
                        print('Registration fee :',data['group1']['Graduate']['S1_2']['ID_3'])
                    elif int(in_num) == 3:
                        print('Registration fee :',data['group1']['Graduate']['S3']['ID_3'])
                    else:print('Invalid semester')


        elif int(in_st[8:10]) in [22,24,26,27,29,34,40,51]: #group2
            if int(in_st[0:2]) in [48,49]: #-------- row1 group2
                if int(in_st[2]) in [3,4]:
                    if int(in_num) in [1,2]:
                        print('Registration fee :',data['group2']['Bachelor']['S1_2']['ID_1'])
                    elif int(in_num) == 3:
                        print('Registration fee :',data['group2']['Bachelor']['S3']['ID_1'])
                    else:print('Invalid semester')
                elif int(in_st[2]) == 7 : 
                    if int(in_num) in [1,2]:
                        print('Registration fee :',data['group2']['Graduate']['S1_2']['ID_1'])
                    elif int(in_num) == 3:
                        print('Registration fee :',data['group2']['Graduate']['S3']['ID_1'])
                    else:print('Invalid semester')

            elif int(in_st[0:2]) in [50,51,52,53,54,55]: # row2 group2
                if int(in_st[2]) in [3,4]:
                    if int(in_num) in [1,2]:
                        print('Registration fee :',data['group2']['Bachelor']['S1_2']['ID_2'])
                    elif int(in_num) == 3:
                        print('Registration fee :',data['group2']['Bachelor']['S3']['ID_2'])
                    else:print('Invalid semester')
                elif int(in_st[2]) == 7 : 
                    if int(in_num) in [1,2]:
                        print('Registration fee :',data['group2']['Graduate']['S1_2']['ID_2'])
                    elif int(in_num) == 3:
                        print('Registration fee :',data['group2']['Graduate']['S3']['ID_2'])
                    else:print('Invalid semester')
            elif int(in_st[0:2]) in [56,57,58,59,60,61,62,63,64,65]: # row3 group2
                if int(in_st[2]) in [3,4]:
                    if int(in_num) in [1,2]:
                        print('Registration fee :',data['group2']['Bachelor']['S1_2']['ID_3'])
                    elif int(in_num) == 3:
                        print('Registration fee :',data['group2']['Bachelor']['S3']['ID_3'])
                    else:print('Invalid semester')
                elif int(in_st[2]) == 7 : 
                    if int(in_num) in [1,2]:
                        print('Registration fee :',data['group2']['Graduate']['S1_2']['ID_3'])
                    elif int(in_num) == 3:
                        print('Registration fee :',data['group2']['Graduate']['S3']['ID_3'])
                    else:print('Invalid semester')

    else:
        print('Invalid ID')
```
